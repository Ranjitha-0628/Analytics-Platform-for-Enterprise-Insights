```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
# Load the CSV dataset
data = pd.read_csv("../datasets/sales_data.csv")

# Show first 5 rows
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data["OrderDate"] = pd.to_datetime(data["OrderDate"])
```

```python
data.dtypes
```

```python
data["Month"] = data["OrderDate"].dt.month
data["Year"] = data["OrderDate"].dt.year
```

```python
plt.figure()
sns.barplot(x="Region", y="Sales", data=data)
plt.title("Total Sales by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.barplot(x="Category", y="Profit", data=data)
plt.title("Profit by Product Category")
plt.show()
```

```python
monthly_sales = data.groupby("Month")["Sales"].sum().reset_index()

plt.figure()
sns.lineplot(x="Month", y="Sales", data=monthly_sales, marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
## Key Business Insights

1. The East and North regions contribute higher average sales, indicating strong market demand.
2. Technology products generate the highest profit margins compared to Furniture and Electronics.
3. Sales show a positive upward trend across months, suggesting business growth and seasonal demand.
```
